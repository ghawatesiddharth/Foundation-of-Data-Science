```python
# Importing required libraries
import matplotlib.pyplot as plt


# Sample data
x = [11,21,32,33,44]
y = [10,20,30,40,50]

# Create a line plot
plt.plot(x, y, color='green', marker='o', linestyle='--', label='Line Plot')

# Adding title and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show legend
plt.grid(True)  # Add grid
plt.show()
```

```python
# Data
categories = ['Sidd', 'Onkar', 'Atharva', 'Chaitu']
marks = [80,90,90,80]
# Create bar plot
plt.bar(categories, marks, color='cyan')

# Add title and labels
plt.title("Marks of Students in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# Sample data
ages = [20,22,24,28,22,20,30,35,26,31,30]

# Create histogram
plt.hist(ages, bins=10, color='orange', edgecolor='black')

# Add title and labels
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

# Create scatter plot
colors = ['blue', 'red', 'green', 'cyan', 'magenta', 'orange', 'purple', 'brown']
plt.scatter(x, y, c=colors, marker='*', s=100)

# Labels and title
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")

# Show plot
plt.show()
```

```python
# Import seaborn
import seaborn as sns
import pandas as pd

# Sample dataset
data = {'Student': ['Sidd', 'Onkar', 'Atharva', 'Chaitu'],
        'Marks': [80, 75, 90, 85]}

df = pd.DataFrame(data)

# Bar plot using seaborn
sns.barplot(x='Student', y='Marks', data=df, palette='Blues_d')
plt.title("Seaborn Bar Plot")
plt.show()
```

```python
# Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temp = [22, 24, 21, 23, 25]

sns.lineplot(x=days, y=temp, marker='o', color='purple')
plt.title("Temperature Over a Week")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
ages = [19,19,20,20,21,22,23,22,22,24,25,30]

# Create histogram
sns.histplot(ages, bins=5, color='limegreen', kde=True)  # kde=True adds smooth curve

# Add title and labels
plt.title("Age Distribution (Seaborn Histogram)")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (correlation-like table)
data = {
    'Math': [80, 75, 90, 85],
    'Physics': [70, 78, 88, 90],
    'Chemistry': [85, 80, 92, 87],
    'Biology': [75, 85, 80, 88]
}

# Convert to DataFrame
df = pd.DataFrame(data, index=['Student A', 'Student B', 'Student C', 'Student D'])

# Create heatmap
sns.heatmap(df, annot=True, cmap='YlGnBu', linewidths=0.5)

# Add title
plt.title("Students Marks Heatmap")
plt.show()
```
